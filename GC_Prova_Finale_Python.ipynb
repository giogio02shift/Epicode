```python
#Notiamo che nel df le colonne continent e location, contengono sia dati puntuali che dati aggregati.
#questo significa che se non dovessimo tenere conto di ciò rischieremmo di contare gli stessi dati più volte.
#i dati aggregati sono per esempio: 
#World, Europe, High income, International (che contengono le somme dei dati di alcuni paesi e che avranno valori  Nan nella colonna continent)
```

```python
## Mantieniamo solo righe con un continente definito ed escludiamo aggregati come 'World', 'Europe', 'High income', ecc.
df_countries = df[df['continent'].notna()].copy()
# Verifica: quanti paesi ci sono
print("Numero di paesi unici:", df_countries['location'].nunique())
```

```python
#Ora vediamo la differenza tra total_cases e new_cases:
#-total-cases, racconta quanti casi in totale ci sono stati finora.
#-new-cases, new_cases racconta quanti casi sono stati confermati oggi.
```

```python
df_countries.shape
```

```python
#il numero di casi per ogni continente fin dall'inizio della pandemia
# Prendi l'ultimo valore cumulativo di total_cases per ogni continente
total_cases_per_continent = df_countries.groupby('continent')['total_cases'].max()

print(total_cases_per_continent)
```

```python
#calcoliamo il totale dei casi covid nel mondo
total_cases_per_continent = df_countries.groupby('continent')['total_cases'].max()
total_cases_world = total_cases_per_continent.sum()
Total_cases_perc = (total_cases_per_continent/total_cases_world)*100
#creiamo un DataFrame con i risultati
world_continent =pd.DataFrame({
    'total_cases' : total_cases_per_continent,
    'total_cases_perc%' : Total_cases_perc.round(2)
})
#ordiniamo per numero di casi in ordine decrescente
world_continent = world_continent.sort_values(by = 'total_cases', ascending = False)
print("il numero totale di casi  nel mondo è", total_cases_world.round(2))
display(world_continent)
```

```python
#Selezionare i dati relativi all'Italia nel 2022 e, poiché i nuovi casi vengono registrati settimanalmente, filtrare via i giorni che non hanno misurazioni; quindi mostrare con dei grafici adatti
# Trasformiamo la colonna date in formato datetime
df_countries['date'] = pd.to_datetime(df_countries['date'])

# Filtra per Italia
df_italy = df_countries[df_countries['location'] == "Italy"]

# Filtra per l'anno 2022
df_italy_2022 = df_italy[df_italy['date'].dt.year == 2022]

# Escludi righe con valori nulli o zero in 'new_cases'
df_italy_2022_clean = df_italy_2022[df_italy_2022['new_cases'].notna() & (df_italy_2022['new_cases'] != 0)]
display(df_italy_2022_clean)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Imposta lo stile di seaborn
sns.set(style="whitegrid")

# Crea la figura
plt.figure(figsize=(12, 6))

# Line plot: nuovi casi per data
sns.lineplot(
    data=df_italy_2022_clean,
    x='date',
    y='new_cases',
    color='firebrick'
)

# Titolo e label
plt.title("Andamento dei nuovi casi COVID-19 in Italia nel 2022", fontsize=16)
plt.xlabel("Data", fontsize=12)
plt.ylabel("Nuovi casi", fontsize=12)

# Asse Y con numeri normali (no scientific notation)
plt.ticklabel_format(style='plain', axis='y')

# Rotazione date asse X
plt.xticks(rotation=45)

# Migliore disposizione layout
plt.tight_layout()
plt.show()
```

```python
#Riguardo le nazioni di Italia, Germania e Francia:
#mostrare in un boxplot la differenza tra queste nazioni riguardo il numero di pazienti in terapia intensiva (Intensive Care Unit, ICU, considerare quindi la colonna icu_patients) da maggio 2022 (incluso) ad aprile 2023 (incluso) 
# scrivere un breve commento (una o due righe) riguardo che conclusioni possiamo trarre osservando il grafico risultante

#filtro per paesi
countries = ['Italy', 'Germany', 'France']
# Filtra per data: da 1 maggio 2022 a 30 aprile 2023
df_filtered = df[(df_countries['location'].isin(countries)) & (df_countries['date'] >= '2022-05-01') & (df_countries['date'] <= '2023/04/30') & (df['icu_patients'].notna())]  # Esclude valori mancanti
display(df_filtered[["icu_patients", "location"]])
# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_filtered, x='location', y='icu_patients')
plt.title('Pazienti in terapia intensiva (ICU)\nMaggio 2022 – Aprile 2023')
plt.xlabel('Paese')
plt.ylabel('Numero di pazienti in ICU')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Nel periodo tra Maggio 2022 ed Aprile 2023  l'italia è stato il paese con meno pazienti in terapia intensiva e la Germania l'opposto.
#pattern e tendenze simili si registrano in Germania e Francia ma gli outiler della Germania sono molto più alti
```
